# applymap with str

```python
people1 = {
   "first": ["hardik", 'mohan', 'randhir'], 
    
    "last": ["kher", 'patel', 'rana'], 
    "email": ["Hardikkher@gmail.com", 'mohanpatel@email.com', 'randhirrana@email.com']
}
```

```python
people = pd.DataFrame(people1)
people
```

```python
people.applymap(len)
```

```python
people.applymap(lambda x : x.upper())
```

```python
people.applymap(lambda x : x.lower())
```

# map() function

```python
people['first'] = people['first'].map({'hardik': 'Anupam', 'randhir': 'ranvir'},na_action=None)
people
```

# Replace method

```python
people = pd.DataFrame(people1)
```

```python
people['first'] = people['first'].replace({'hardik': 'Anupam', 'randhir': 'ranvir'})
people
```
