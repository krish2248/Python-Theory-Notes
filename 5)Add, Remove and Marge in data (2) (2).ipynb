# marge two dataset with on and how

```python
df1 = pd.DataFrame({"fruit" : ["apple", "banana", "orange","grape"],
                    "market_price" : [100, 40, 80, 60]})
df1
```

```python
df2 = pd.DataFrame({"fruit" : ["apple", "banana", "grape",'kiwi'],
                    "wholesaler_price" : [60, 20, 45,100]})
df2
```

```python
#  Merging two Dataframe with different number of elements :
```

```python
pd.merge(df1,df2,how='inner',on='fruit')
```

```python
pd.merge(df1,df2,how='right',on='fruit')
```

```python
pd.merge(df1,df2,how='left',on='fruit')
```

```python
pd.merge(df1,df2,how='outer',on='fruit')
```

```python
pd.merge(df1,df2,how='outer',on='fruit',indicator=True)
```

```python
df1 = pd.DataFrame({"fruit" : ["apple", "banana","orange", "grape"],
                    "Vagitables" : ['cucumber','tometo','poteto','carrot'],
                    "market_price" : [100, 40, 80, 60]})
df1
```

```python
df2 = pd.DataFrame({"fruit" : ["apple", "banana", "grape",'dragon_fruit'],
                    "Vagitables" : ['cucumber','tometo','cabbige','carrot'],
                    "wholesaler_price" : [60, 20, 45,100]})
df2
```

```python
pd.merge(df1,df2 , how= 'inner' , on =['Vagitables', 'fruit'])
```

```python
pd.merge(df1,df2 , how= 'inner')    # by defualt take all colums
```

```python
pd.merge(df1,df2 , how= 'inner',on='fruit')
```

```python
df3 = pd.DataFrame({"ID" : [1,2,3,4],
                    "Name" : ["a","b","c","d"]})
df3
```

```python
df4 = pd.DataFrame({"ID" : [5,6,7,8],
                    "Name" : ["a","b","c","d"]})
df4
```

```python
pd.merge(df3,df4,left_index=True,right_index=True)   # use left_index & right_index use together
```

```python
pd.merge(df3,df4,left_index=True,right_index=True,suffixes=["_df3","_df4"])
```

# concat()

```python
# concate with axis = 0,1
```

```python
people1 = {
    "first": ["hardik", 'harshil', 'dharuv'], 
    "last": ["kher", 'vaghela', 'giri'], 
    }
```

```python
people2 = {
    "first": ["ram", 'shayam', 'shiv'], 
    "last": ["kher", 'vaghela', 'giri'], 
    }
```

```python
df1 = pd.DataFrame(people1)
df1
```

```python
df2 = pd.DataFrame(people2)
df2
```

```python
df3 = pd.concat([df1,df2])     # default  axis = 0
df3
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df3 = pd.concat([df1,df2],axis=0,ignore_index=True)
df3
```

```python
df3 = pd.concat([df1,df2],axis=1,ignore_index=True)
df3
```

```python
people2 = {
    "first": ["ram", 'shayam'], 
    "last": ["kher", 'vaghela'], 
    }
new = pd.DataFrame(people2)
new
```

```python
pd.concat([df1,new],axis=1,join = "inner")
```

```python
pd.concat([df1,df2],keys = ["First","Second"])
```

# join()

```python
# join two dataset also try with different index
```

```python
df1 = pd.DataFrame({'ID1':[1,2,3,4,5],'class':[11,22,33,44,55]})
df1
```

```python
df2 = pd.DataFrame({'ID2':[6,7,8,9,10], 'Name':['hardik','dharuv','harshil','ram','shyam']})
df2
```

```python
df1.join(df2)     # index must be same.
```

```python
df2.join(df1)
```

```python
df3 = pd.DataFrame({'ID2':[6,7,8], 'Name':['hardik','dharuv','harshil']})
df3
```

```python
df1.join(df3)
```

```python
df1.join(df3,how="inner")
```

```python
df4 = pd.DataFrame({'ID1_':[6,7,8,9,10], 'Name':['hardik','dharuv','harshil','ram','shyam']})
df4
```

```python
df1.join(df4)
```

```python
df1.join(df4,lsuffix='_4545')
```

```python
df1.join(df4,rsuffix='_7777')
```

```python
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=[0, 1, 2])
df4 = pd.DataFrame({'B': [7, 8, 9], 'C': [10, 11, 12]}, index=[0, 1, 2])

print(df1)
print(df4)
```

```python
result = df1.join(df4, rsuffix='_7777')
print(result)
```
