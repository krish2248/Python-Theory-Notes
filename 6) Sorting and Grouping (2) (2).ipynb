```python
import pandas as pd
import os

path = r"D:\python lerning\Dataset\survey_results_public.csv"

if os.path.exists(path):
    print("File found! Reading data...\n")
    df = pd.read_csv(path)
else:
    print("File not found. Check the path.")
    exit()

print("\nDataset Info:")
print(df.info())

print("\nColumns in dataset:", df.columns)

column_name = "Country" if "Country" in df.columns else "country"

print(f"\nData type of '{column_name}':", df[column_name].dtype)

if not pd.api.types.is_numeric_dtype(df[column_name]):
    print("\n'country' column contains non-numeric data. Counting occurrences instead:")
    print(df[column_name].value_counts())
else:
    df[column_name] = pd.to_numeric(df[column_name], errors="coerce")
    print("\nMean of 'country' column:", df[column_name].mean())

if "some_numeric_column" in df.columns:
    print("\nMean of 'some_numeric_column' grouped by 'country':")
    print(df.groupby(column_name)["some_numeric_column"].mean())
else:
    print("\nColumn 'some_numeric_column' does not exist in the dataset.")
```
