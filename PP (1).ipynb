```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner  # Public attribute
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance: {self.__balance}")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn {amount}. New balance: {self.__balance}")
        else:
            print("Invalid withdrawal amount or insufficient balance.")

    def get_balance(self):  # Public method to access private attribute
        return self.__balance

    def set_balance(self, new_balance):  # Public method to modify private attribute
        if new_balance >= 0:
            self.__balance = new_balance
            print(f"Balance updated to {self.__balance}")
        else:
            print("Invalid balance value.")

# Taking user input
owner = input("Enter account owner name: ")
balance = float(input("Enter initial balance: "))

account = BankAccount(owner, balance)

while True:
    print("\n1. Deposit\n2. Withdraw\n3. Check Balance\n4. Update Balance (Admin)\n5. Exit")
    choice = input("Choose an option: ")
    
    if choice == '1':
        amount = float(input("Enter deposit amount: "))
        account.deposit(amount)
    elif choice == '2':
        amount = float(input("Enter withdrawal amount: "))
        account.withdraw(amount)
    elif choice == '3':
        print(f"Current balance: {account.get_balance()}")
    elif choice == '4':
        new_balance = float(input("Enter new balance: "))
        account.set_balance(new_balance)
    elif choice == '5':
        print("Exiting...")
        break
    else:
        print("Invalid choice. Try again.")
```

```python
class DynamicAttribute:
    def __init__(self, name, value, private=False):
        self.name = name
        if private:
            self.__value = value  # Private attribute
        else:
            self.value = value  # Public attribute
        self.is_private = private

    def get_value(self):
        if self.is_private:
            return self.__value
        return self.value

    def set_value(self, new_value):
        if self.is_private:
            self.__value = new_value
        else:
            self.value = new_value

# Taking user input
name = input("Enter attribute name: ")
value = input("Enter attribute value: ")
private = input("Should this be private? (yes/no): ").strip().lower() == "yes"

attribute = DynamicAttribute(name, value, private)

while True:
    print("\n1. Get Value\n2. Set Value\n3. Exit")
    choice = input("Choose an option: ")
    
    if choice == '1':
        print(f"{attribute.name}: {attribute.get_value()}")
    elif choice == '2':
        new_value = input("Enter new value: ")
        attribute.set_value(new_value)
        print("Value updated successfully.")
    elif choice == '3':
        print("Exiting...")
        break
    else:
        print("Invalid choice. Try again.")
```

```python
class DynamicAttribute:
    def __init__(self, name, value, private=False):
        self.name = name
        if private:
            self.__value = value  # Private attribute
        else:
            self.value = value  # Public attribute
        self.is_private = private

    def get_value(self):
        if self.is_private:
            return self.__value
        return self.value

    def set_value(self, new_value):
        if self.is_private:
            self.__value = new_value
        else:
            self.value = new_value

    def access_private_directly(self):
        return getattr(self, f"_{self.__class__.__name__}__value", "Not a private attribute")

    def delete_private_data(self):
        if self.is_private:
            self.__value = None
            print("Private data deleted!")

# Taking user input
name = input("Enter attribute name: ")
value = input("Enter attribute value: ")
private = input("Should this be private? (yes/no): ").strip().lower() == "yes"

attribute = DynamicAttribute(name, value, private)

password = "securepass"  # Set a predefined password
user_password = input("Enter password to access private data: ")

if user_password != password:
    print("Wrong password! Deleting private data...")
    attribute.delete_private_data()

while True:
    print("\n1. Get Value\n2. Set Value\n3. Access Private Directly\n4. Exit")
    choice = input("Choose an option: ")
    
    if choice == '1':
        print(f"{attribute.name}: {attribute.get_value()}")
    elif choice == '2':
        new_value = input("Enter new value: ")
        attribute.set_value(new_value)
        print("Value updated successfully.")
    elif choice == '3':
        print(f"Direct Access: {attribute.access_private_directly()}")
    elif choice == '4':
        print("Exiting...")
        break
    else:
        print("Invalid choice. Try again.")
```
